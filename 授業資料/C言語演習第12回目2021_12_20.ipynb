# §1. 定型処理は関数におまかせ

※ 引数（ひきすう）と返り値という用語は思い出しておこう。

今までもいろいろな関数を使ってきました。

* printf, scanf, puts
* sqrt, sin, acos

これらの関数はいちいち書くと面倒な処理や難しい処理をやってくれるのでした。

sin 関数などは専門書などのコードを参考にして作成することもできますが、必要になるたびに調べるのは手間ですし、ミスも起こりやすくなります。
こういった定型処理は関数化しておくと、手間が省けるだけでなく、ミスも防げますし、さらに、より複雑な処理も見通し良く記述できます。

コード

```python
%%file 12hypot.c
#include <stdio.h>
#include <math.h>

int main(void)
{
    double a, b;
    double c;
 
    printf("直角三角形の底辺と高さから斜辺の長さを計算します。\n");
    printf("底辺： "); scanf("%lf", &a);
    printf("高さ： "); scanf("%lf", &b);
 
    c = hypot(a, b);  // もちろん、c = sqrt(a * a + b * b); と書いてもいいです。
    // a, b が引数（ひきすう）で、関数 hypot による計算結果を返り値といい、それが hypot(a, b) の値となります。 
 
    printf("底辺 %.2f、高さ %.2f の直角三角形の斜辺の長さは %.2f です。\n", a, b, c);

    return 0;
}
```

# §3. 仮引数（かりひきすう）と実引数（じつひきすう）



仮引数というのは、数学でいうと、パラメータを持つ関数の「パラメータ」に近い感じです。
たとえば
$$y = ax +b$$
という $x$ の関数はパラメータ $a$、$b$ を持っています。
値 $x$ についてのこの関数の値 $y$ は、これらパラメータの値が決まらないと求められません。

これと同様な状況が bmi 関数にあります。
すなわち、height と weight が決まらないと結果が計算できません。
height と weight のような引数を **仮引数(かりひきすう）** といいます（図を参照せよ）。

それに対して、main 関数内の `bmi(h, w)` の部分においては、ここを通過する際は h、w の値は決まっています。
これら h、w のような引数を **実引数（じつひきすう）** といいます。

スクリーンショット 2021-12-16 18.26.52.png

```python
%%file 12bmi2.c
#include <stdio.h>

double bmi(double height, double weight)
{
    double b;
 
    b =  weight / ((height / 100) * (height / 100));

    return b;
}

int main(void)
{
    double h, w;

    printf("身長[cm]: "); scanf("%lf", &h);
    printf("体重[kg]: "); scanf("%lf", &w);
  
    printf("BMI は %.1f です。\n", bmi(h, w));

    return 0;
}
```

コンパイルと実行

```python
!gcc 12bmi2.c && ./a.out
```

## 5.2. 3 のつく数

コード

※ 関数名は san_tsuku としてください。（関数名や変数名の1文字目に数字を用いることはできません。）

```python
%%file 12nabeatu.c
#include <stdio.h>

// ここに san_tsuku 関数を記述

int main(void)
{
    int n, result;
    int i;
 
    // 入力部分
    printf("n = "); scanf("%d", &n);

    // 主要処理部分
    result = 0;
    while (n > 0)
    {
        if (n % 10 == 3)
        {
            result = 1;
            break;
        }
        n = n / 10;
    }
 
    // 結果表示部分
    // 関数化前は結果表示がおかしいですが、関数化すると直ります。
    if (result == 0)
    {
        printf("%d は 3 のつかない数です。\n", n);
    }
    else
    {
        printf("%d は 3 のつく数です。\n", n);
    }
 
    return 0;
}
```

## 5.3. n x n の碁盤の目を描く → n x m の碁盤の目を描く

以下のプログラムは n を入力させて、n x n の碁盤の目を描くもので、すでに関数化されています。

これを **n と m を入力させて、n x m の碁盤の目を描くように変更** してください。

コード

```python
%%file 12grid.c
#include <stdio.h>

/**
 * 碁盤の目を描く関数
 *
 * 入力 int n .... 縦方向の目の個数（= 横方向の目の個数）
 * 出力 void  .... （なし）
 */
void draw_grid(int n)
{
    int i, j;

    printf("┏");
    for (j = 1; j < n; j++)
    {
        printf("┯");
    }
    printf("┓\n");
 
    for (i = 1; i < n; i++)
    {
        printf("┠");
        for (j = 1; j < n; j++)
        {
            printf("┼");
        }
        printf("┨\n");
    }

    printf("┗");
    for (j = 1; j < n; j++)
    {
        printf("┷");
    }
    printf("┛\n");
}

int main(void)
{
    int n;
 
    printf("サイズ： "); scanf("%d", &n);
    draw_grid(n);

    return 0;
}
```

# §6. 関数プロトタイプ宣言



複雑な処理をするようになると、自作関数が増えたり関数の行数が増えたりします。こうなると main 関数がどこか分かりにくくなり、メンテナンスもしにくくなります。

そういう場合、コードを複数ファイル化し別々にコンパイルして、あとで結合するということをやるのですが、その際に関数の引数や返り値の情報はあらかじめコンパイラが知っておく必要があります。
その情報を記述するのが、関数のプロトタイプ宣言です。（もちろん、関数の処理部本体は別途記述する必要があります。）

ちなみに、ヘッダファイル（`stdio.h`、`math.h` など）にはこういったプロトタイプ宣言などが記述されています。

ここで例示するコードでは、関数プロトタイプ宣言をコード冒頭に置き、関数の処理部本体はソースファイルの末尾に記述しています。
あまり長いプログラムではありませんが、これでもある程度は見やすくなったのではないでしょうか。（碁盤の目のプログラムだと、もっとその効果がわかりやすいかもしれません。）

コード（bmi 関数のプロトタイプ宣言をしないと動きません）

```python
%%file 12bmi6.c
#include <stdio.h>

// ここに bmi 関数のプロトタイプを記述する必要あり。

int main(void)
{
    double h, w;

    printf("身長[cm]: "); scanf("%lf", &h);
    printf("体重[kg]: "); scanf("%lf", &w);
  
    printf("BMI は %.1f です。\n", bmi(h, w));

    return 0;
}

double bmi(double height, double weight)
{
    double b;
 
    b =  weight / ((height / 100) * (height / 100));

    return b;
}
```
